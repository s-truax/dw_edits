```python
import csv
```

```python
deck_export = 'DW_Nicos_Weg_A1.txt'
```

```python
deck_data = []
```

```python
with open(deck_export, encoding='utf-8', newline='') as f:
    reader = csv.reader(f, delimiter='\t')
    for _ in range(3):
        next(reader)  # skip metadata
    for row in reader:
        deck_data.append(row)
```
